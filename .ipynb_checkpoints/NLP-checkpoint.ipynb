```python
!pip install nltk
!pip install scikit-learn

import nltk

# Download required files (run once)
nltk.download('punkt_tab')
nltk.download('stopwords_tab')
nltk.download('wordnet_tab')
nltk.download('vader_lexicon_tab')

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
text = """
I love Python programming.
Python is very easy to learn.
I enjoy coding every day.
Sometimes debugging is difficult.
"""

print("Original Text:\n")
print(text)
```

```python
print("\nWord Tokenization:\n")
words = word_tokenize(text)
print(words)
```

```python
print("\nSentence Tokenization:\n")
sentences = sent_tokenize(text)
print(sentences)
```

```python
# Stop Words Removal
print("\nStop Words Removal:\n")
stop_words = set(stopwords.words('english'))

# Custom stop word
stop_words.add("Python")

filtered_words = [word for word in words
                  if word.lower()
                  not in stop_words and word.isalpha()]
print(filtered_words)
```

```python
# Stemming
print("\nStemming:\n")
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_words]
print(stemmed_words)
```

```python
print("\nLemmatization:\n")
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print(lemmatized_words)
```

```python
print("\ncountvectorizer:\n")
documents=[
    "I love python Progamming",
    "Python is easy to learn",
    "I enjoy coding"
]
cv=CountVectorizer()
count_matrix=cv.fit_transform(documents)
print(cv.get_feature_names_out())
print(count_matrix.toarray())  # here this matrix is maping with each index of document and finding count 
```

```python
print("\nTfidfVectorizer:\n")
tdf=TfidfVectorizer()
tdf_matrix=tdf.fit_transform(documents)
print(tdf.get_feature_names_out())
print(tdf_matrix.toarray())
```
