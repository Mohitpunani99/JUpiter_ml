```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
X=np.array([1,2,3,4,5])
Y=np.array([2,4,5,4,5])
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)
print(f" The mean of x is={mean_x}")
print(f" The mean of x is={mean_y}")
```

```python
num=np.sum((X-mean_x)*(Y-mean_y))
deno=np.sum((X-mean_x)**2)
slope=num/deno
print(slope)
```

```python
intercept=mean_y-slope*mean_x
print(intercept)
```

```python
# X=int(input())
Y_pread=(slope*X)+intercept
print(Y_pread)
```

```python
# linear regression by model
```

```python
x_reshaped=X.reshape(-1,1)
print(x_reshaped)
model=LinearRegression()
model.fit(x_reshaped,Y)
```

```python
slope_by_lr=model.coef_[0]
interceplt_by_lr=model.intercept_
print(interceplt_by_lr)
```

```python
target_x=6
y_pread_LR=model.predict([[target_x]])
print(y_pread_LR)
```

```python
plt.scatter(X,Y,color="blue",label="Actual data")
plt.scatter(X,Y_pread,color="green",label="predicted data")
plt.plot(X,Y_pread,color="red",label="Regression Line") #ploting the regression line for predicted points
plt.xlabel("study hours (x)")
plt.ylabel("exan score (y)")
plt.legend()
plt.show() # showing data 
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error

r2_score=r2_score(Y,Y_pread)
print(f"r2 score is {r2_score:.2f}")

mse=mean_squared_error(Y,Y_pread)
print(f"mean_squared_error is {mse:.2f}")

mae=mean_absolute_error(Y,Y_pread)
print(f"mean_absolute_error is {mae:.2f}")

rmse=root_mean_squared_error(Y,Y_pread)
print(f"root_mean_squared_error is {rmse:.2f}")
```

```python
# if r2 score is 1 then input and target varriable over lap each other
# which is not good
```

```python
# using pandas to make a DF

data={
    "Area":[500,800,1000,1200,1500,1800],
    "price":[50,80,100,120,150,160]
}
df=pd.DataFrame(data)
```

```python
df.head()
```

```python

x=df[["Area"]] # it is 2d aray 
y=df["price"] 
# A_reshaped=A.reshape(-1,1)
# print(A_reshaped)
model=LinearRegression()
model.fit(x,y)

slope_by_lr=model.coef_[0]
interceplt_by_lr=model.intercept_
print(interceplt_by_lr)


y_pred_LR=model.predict(x)
print(y_pred_LR)



plt.scatter(x,y,color="blue",label="Actual data")
plt.scatter(x,y_pred_LR,color="green",label="predicted data")
plt.plot(x,y_pred_LR,color="red",label="Regression Line") #ploting the regression line for predicted points
plt.xlabel("Area  (x)")
plt.ylabel(" price (y)")
plt.legend()
plt.show() # showing data 
```
