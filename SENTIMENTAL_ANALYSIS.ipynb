```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Restaurant_reviews.tsv",delimiter="\t",quoting=3)
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

def rep(review):
    # Remove non-alphabetic characters
    review = re.sub('[^a-zA-Z]', ' ', review)
    
    # Convert to lowercase
    review = review.lower()
    
    # Split into words
    reviews = review.split()
    
    # Remove stopwords and apply stemming
    reviews = [stemmer.stem(x) for x in reviews
               if x not in stopwords.words('english')]
    
    # Join words back into a string
    review = ' '.join(reviews)
    
    return review
```

```python
df["Review"]=df["Review"].apply(rep)
df.head()
```

```python
# corpus = CountVectorizer(max_features=1500)
# x = cv.fit_transform(corpus).toarray()
# y = df["liked"].tolist()=df["Review"].tolist()
# corpus[:5]

from sklearn.feature_extraction.text import CountVectorizer

corpus = df["Review"].tolist()   # your cleaned text, as a list of strings

cv = CountVectorizer(max_features=1500)
x = cv.fit_transform(corpus).toarray()
y = df["Liked"].tolist()         
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,Normalizer,MinMaxScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf=MultinomialNB()
clf.fit(x_train,y_train)
print("Accuracy:{0:.if}%".
     format(clf.score(x_test,y_test)*100))
```

```python
def predict_sentiment(review_text):
    # Preprocess text using your existing function
    cleaned_review = rep(review_text)
    
    # Convert text to vector
    vector = cv.transform([cleaned_review]).toarray()
    
    # Predict
    prediction = clf.predict(vector)[0]
    
    # Result
    if prediction == 1:
        return "Positive Sentiment"
    else:
        return "Negative Sentiment"


# Example predictions
review1 = "The food was amazing and staff was very friendly"
review2 = "Worst restaurant ever, food was cold and tasteless"

print(predict_sentiment(review1))
print(predict_sentiment(review2))
```
