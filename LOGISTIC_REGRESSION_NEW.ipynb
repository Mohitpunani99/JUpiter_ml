```python
import  pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=np.array([[1],[2],[3],[4],[5]])
y=np.array([0,0,0,1,1])
```

```python
model=LogisticRegression()
model.fit(x,y)
```

```python
x_test=np.linspace(0,6,10).reshape(-1,1) # it create line interval bw line to gether data 
prob=model.predict_proba(x_test)
 # print(prob)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
```

```python
plt.scatter(x,y,color="red")
plt.scatter(x_test,model.predict(x_test),color="yellow",label="Class")
plt.scatter(x_test,prob[:,0],color="blue")
plt.scatter(x_test,prob[:,1],color="black")
plt.plot(x_test, prob, color="green", label="sigmoid curve")
plt.xlabel("Study Hours")
plt.ylabel("prob")
plt.legend()
plt.show()
```
