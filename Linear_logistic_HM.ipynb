# House Price Predication 

```python
import  pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics  import accuracy_score , confusion_matrix ,classification_report
```

```python
df=pd.read_csv("housedata.csv")
df.head()
```

```python
x=df[["sqft_living"]]
y=df["price"]
```

```python
# x1=df[["sqft_living"]].max()
# x1
```

```python

# A_reshaped=A.reshape(-1,1)
# print(A_reshaped)
model=LinearRegression()
model.fit(x,y)

slope_by_lr=model.coef_[0]
interceplt_by_lr=model.intercept_
print(interceplt_by_lr)


y_pred_LR=model.predict(x)
print(y_pred_LR)



plt.scatter(x,y,color="blue",label="Actual data")
plt.scatter(x,y_pred_LR,color="green",label="predicted data")
plt.plot(x,y_pred_LR,color="red",label="Regression Line") #ploting the regression line for predicted points
plt.xlabel("sqft_living(x)")
plt.ylabel(" price (y)")
plt.legend()
plt.show() # showing data 
```

# Age Estimated Prediction 

```python
pf=pd.read_csv("social.csv")
pf.head()
```

```python
X=pf[["Age"]]
Y=pf["EstimatedSalary"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42) 
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)

slope_by_lr=model.coef_[0]
interceplt_by_lr=model.intercept_
print(interceplt_by_lr)


y_pred_LR=model.predict(X_test)
print(y_pred_LR)



# plt.scatter(X_train,Y_train,color="blue",label="Actual data")
# plt.scatter(X,y_pred_LR,color="green",label="predicted data")
# plt.plot(X,y_pred_LR,color="red",label="Regression Line") #ploting the regression line for predicted points
# plt.xlabel("Age(x)")
# plt.ylabel("EstimatedSalary")
# plt.legend()
# plt.show() # showing data 

plt.scatter(X_train, Y_train, color="blue", label="Train data")
plt.scatter(X_test, Y_test, color="orange", label="Test Actual")
plt.scatter(X_test, y_pred_LR, color="green", label="Predicted")
plt.plot(X_test, y_pred_LR, color="red", label="Regression Line")
plt.xlabel("Age (X)")
plt.ylabel("Estimated Salary")
plt.legend()
plt.show()
```

```python
dl=pd.read_csv("social.csv")
dl.head()
```

```python
u=dl[["Age"]]
v=dl["Purchased"]
```

```python
u_train,u_test,v_train,v_test=train_test_split(u,v,test_size=0.25,random_state=42)
model=LogisticRegression()
model.fit(u_train,v_train)
```

```python
# x_test=np.linspace(0,6,10).reshape(-1,1) # it create line interval bw line to gether data 
```

```python
y_pred=model.predict(u_test)
print(y_pred)
```

```python
prob=model.predict_proba(u_test)
```

```python
plt.scatter(u, v, color="red")
plt.scatter(u_test, model.predict(u_test), color="yellow", label="Class")
plt.scatter(u_test, prob[:,0], color="blue")
plt.scatter(u_test, prob[:,1], color="black")
plt.plot(u_test, prob, color="green", label="sigmoid curve")
plt.xlabel("AGE")
plt.ylabel("Prob")
plt.legend()
plt.show()
```

```python
acc=accuracy_score(v_test,y_pred)
print(f"the accuracy is :{acc}")
cr=classification_report(v_test,y_pred)
print(cr)
```
